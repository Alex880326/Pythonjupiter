```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

#### <span style="color: Salmon">Тепловая карта </span>

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Этажность')
plt.title('Зависимость цен от этажей');
```

#### <span style="color: Salmon">Дома, у которых  2.5 этажа, стоят дороже </span> 

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
plt.xlabel('Стоимость')
plt.ylabel('Год постройки');
```

#### <span style="color: Salmon">Год постройки влияет на цену</span>

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
plt.xlabel('Стоимость')
plt.ylabel('Дизайн и качество постройки');
```

#### <span style="color: Salmon">Дизайн и качество постройки также сказываются на стоимости</span>

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Обзор')
plt.title('Оценка по общему виду');
```

#### <span style="color: Salmon">Хороший вид увеличивает стоимость домов</span>
